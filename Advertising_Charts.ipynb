```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a synthetic dataset
np.random.seed(0)
dates = pd.date_range(start='2022-01-01', end='2022-12-31', freq='M')
channels = ['TV', 'Online', 'Social Media', 'Outdoor']
data = {channel: np.random.randint(low=100, high=1000, size=len(dates)) for channel in channels}
df = pd.DataFrame(data, index=dates)

# Plot the data
df.plot(kind='barh')
plt.title('Leads Generated from Different Channels Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Leads')
plt.legend(title='Channel')
plt.show()
```

```python
# Create a synthetic dataset for advertising spend
spend = {channel: np.random.randint(low=4000, high=25000, size=len(dates)) for channel in channels}
df_spend = pd.DataFrame(spend, index=dates)

# Calculate ROAS
roas = df.div(df_spend) * 300  # Assuming revenue per lead is $300

# Plot the average ROAS by channel
roas.mean().plot(kind='barh')
plt.title('Average ROAS by Channel')
plt.xlabel('Channel')
plt.ylabel('ROAS')
plt.show()
```

```python
import seaborn as sns
# Prepare data for a heatmap
df['Month'] = df.index.month_name()
df_melt = df.melt(id_vars='Month', var_name='Channel', value_name='Leads')

# Create a pivot table for the heatmap
pivot = df_melt.pivot_table(index='Month', columns='Channel', values='Leads')

# Create a heatmap
sns.heatmap(pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap of Leads by Channel and Month')
plt.xlabel('Channel')
plt.ylabel('Month')
plt.show()
```

RFM (Recency, Frequency, Monetary) analysis is a customer segmentation technique that uses past purchase behavior to divide customers into groups. It is especially common in the retail sector but can also be applied in the context of advertising to understand and segment audiences. For simplicity, let's assume that we can translate these concepts into:

Recency: How recently a customer has interacted with an ad.
Frequency: How often a customer interacts with ads.
Monetary: The customer's average session duration.

```python
# Creating a synthetic dataset
np.random.seed(0)
n_customers = 500
recency = np.random.randint(low=1, high=100, size=n_customers)  # days since last ad interaction
frequency = np.random.randint(low=1, high=10, size=n_customers)  # ad interactions in the last month
monetary = np.random.randint(low=1, high=1000, size=n_customers)  # average session duration in seconds

df_rfm = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'Monetary': monetary})

# Plotting
sns.pairplot(df_rfm)
plt.suptitle('RFM Distributions and Relationships', y=1.02)
plt.show()
```

Sankey diagram illustrating ad funnel and tracking: impressions, clicks, conversions, drop off after impressions and drop off after clicks.

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["Impressions", "Clicks", "Conversions", "Dropoff after Impressions", "Dropoff after Clicks"],
      color = ["blue", "blue", "blue", "red", "red"]
    ),
    link = dict(
      source = [0, 0, 1, 1], # indices correspond to labels, eg Impression -> Clicks, Impression -> Dropoff after Impression, Clicks -> Conversions, Clicks -> Dropoff after Clicks
      target = [1, 3, 2, 4],
      value = [200, 800, 50, 150]
  ))])

fig.update_layout(title_text="Ad Conversion Funnel", font_size=10)
fig.show()
```

```python
import pandas as pd
import numpy as np

np.random.seed(0)  # For reproducibility

# Define the data for each marketing channel
data = {
    "Search": {
        "Targeting": np.random.choice(['Broad', 'Narrow'], size=100),
        "Impressions": np.random.randint(1000, 5000, size=100),
        "Clicks": np.random.randint(100, 500, size=100),
        "Conversions": np.random.randint(10, 100, size=100),
        "Cost": np.random.uniform(100.0, 500.0, size=100)
    },
    "Social": {
        "Targeting": np.random.choice(['Broad', 'Narrow'], size=100),
        "Impressions": np.random.randint(5000, 10000, size=100),
        "Clicks": np.random.randint(200, 1000, size=100),
        "Conversions": np.random.randint(20, 200, size=100),
        "Cost": np.random.uniform(500.0, 1000.0, size=100)
    },
    "Email": {
        "Targeting": np.random.choice(['Broad', 'Narrow'], size=100),
        "Impressions": np.random.randint(10000, 15000, size=100),
        "Clicks": np.random.randint(500, 1500, size=100),
        "Conversions": np.random.randint(50, 300, size=100),
        "Cost": np.random.uniform(1000.0, 1500.0, size=100)
    },
    "Display": {
        "Targeting": np.random.choice(['Broad', 'Narrow'], size=100),
        "Impressions": np.random.randint(15000, 20000, size=100),
        "Clicks": np.random.randint(1000, 2000, size=100),
        "Conversions": np.random.randint(100, 400, size=100),
        "Cost": np.random.uniform(1500.0, 2000.0, size=100)
    }
}

# Create a dataframe for each marketing channel
dfs = {channel: pd.DataFrame(channel_data) for channel, channel_data in data.items()}

# Display the dataframes
for channel, df in dfs.items():
    print(f"{channel} Data:")
    print(df.head())
    print("\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# For each marketing channel
for channel, df in dfs.items():
    # Create a scatter plot matrix
    sns.set(style="ticks")
    sns.pairplot(df, hue="Targeting")

    # Display the plot
    plt.suptitle(f"{channel} Scatter Plot Matrix", y=1.02)
    plt.show()
```

```python
# Calculate the cost per conversion for each channel
for channel, df in dfs.items():
    df['Cost per Conversion'] = df['Cost'] / df['Conversions']

# Combine the dataframes
combined_df = pd.concat(dfs, names=['Channel', 'Row'])

# Reset the index so that 'Channel' becomes a column
combined_df.reset_index(level=0, inplace=True)

# Create a bar chart of cost per conversion by channel
plt.figure(figsize=(10,6))
sns.barplot(x='Channel', y='Cost per Conversion', data=combined_df)
plt.title('Cost per Conversion by Channel')
plt.show()
```

```python
# Calculate the CTR for each row in each dataframe
for df in dfs.values():
    df['CTR'] = df['Clicks'] / df['Impressions']

# Create a boxplot of CTR by targeting type for each channel
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('CTR by Targeting Type', y=0.92)

for ax, (channel, df) in zip(axs.flatten(), dfs.items()):
    sns.boxplot(x='Targeting', y='CTR', data=df, ax=ax)
    ax.set_title(channel)

plt.show()
```
