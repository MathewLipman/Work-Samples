X_coords can be derived using the ax.get_xticks () and ax.get_yticks() to get array of x and y axis coordinates for placing text using ax.text() . Its lot of trial and error but it gets better as you try it few times. Please let me know if you end up finding any better efficient way to manage placing text.

```python
style.use('dark_background')

# initilizes our figure and chart
fig, ax = plt.subplots(figsize=(9, 5))
# designate chart types bar, data source, positions and size
ax.barh(white_corr.index, white_corr, left=2, height=0.5, color='white')
ax.barh(red_corr.index, red_corr, height=0.5, left=-0.1, color='red')

#remove grid lines and xtick and ytick labels
ax.grid(visible=None, which='major')
ax.set_yticklabels([])
ax.set_xticklabels([])

#set label
x_coords = {'Alcohol': 0.82, 'Sulphates': 0.77, 'pH': 0.91,
            'Density': 0.80, 'Total Sulfur Dioxide': 0.59,
            'Free Sulfur Dioxide': 0.6, 'Chlorides': 0.77,
            'Residual Sugar': 0.67, 'Citric Acid': 0.76,
            'Volatile Acidity': 0.67, 'Fixed Acidity': 0.71}
y_coord = 9.8
for y_label, x_coord in x_coords.items():
    ax.text(x_coord, y_coord, y_label)
    y_coord -= 1
    
ax.axvline(0.5, c='grey', alpha=0.1, linewidth=1,
           ymin=0.1, ymax=0.9)
ax.axvline(1.45, c='grey', alpha=0.1, linewidth=1,
           ymin=0.1, ymax=0.9)

ax.axhline(-1, color='grey', linewidth=1, alpha=0.5,
          xmin=0.01, xmax=0.32)


ax.text(-0.7, -1.7, '-0.5'+ ' '*31 + '+0.5',
        color='grey', alpha=0.5)
ax.axhline(-1, color='grey', linewidth=1, alpha=0.5,
           xmin=0.67, xmax=0.98)
ax.text(1.43, -1.7, '-0.5'+ ' '*31 + '+0.5',
        color='grey', alpha=0.5)

ax.axhline(11, color='grey', linewidth=1, alpha=0.5,
          xmin=0.01, xmax=0.32)
ax.text(-0.33, 11.2, 'RED WINE', weight='bold', color='red')

ax.axhline(11, color='grey', linewidth=1, alpha=0.5,
          xmin=0.67, xmax=0.98)
ax.text(1.75, 11.2, 'WHITE WINE', weight='bold', color='white')

ax.text(-0.7, -2.9,
        'Mathew Lipman',
        color = '#f0f0f0',
        backgroundcolor = '#4d4d4d',
        size=12)

ax.text(-0.7, 13.5,
        'Wine Quality Most Strongly Correlated With Alcohol Level',
        size=17, weight='bold')
ax.text(-0.7, 12.7,
        'Correlation values between wine quality and wine properties (alcohol, pH, etc.)')

plt.show()
```
