```python
import matplotlib.pyplot as plt
```

The plt.plot() function generates a line graph by default. All it needs is two arrays of data of the same length â€” these can be Python lists, pandas Series, NumPy arrays, etc. Above, we used two Python lists.

Notice the order of arguments in plt.plot(month_number, new_cases): month_number comes first, followed by new_cases. The array that comes first gives the x-coordinates, and the second array gives the y-coordinates.

```python
month_number = [1, 2, 3, 4, 5, 6, 7]
new_cases = [9926, 76246, 681488, 2336640,
             2835147, 4226655, 6942042]

plt.plot(month_number, new_cases)
plt.show()
```

The le6 is telling us in scientific notation that the values on y-axis are multiplied 10^6 

adding the plt.ticklabel_format style='plain' for the y axis allows us to r to remove the scientific notation on the y axis and get a better sense of the scale. 

```python


month_number = [1, 2, 3, 4, 5, 6, 7]
new_deaths = [213, 2729, 37718, 184064, 143119, 136073, 165003]

plt.plot(month_number, new_deaths)
plt.ticklabel_format(axis='y', style='plain')
```

now we went ahead and labeled our corresponding axis with xlabel and ylabel methods. We also added a title using title method.

```python
month_number = [1, 2, 3, 4, 5, 6, 7]
new_deaths = [213, 2729, 37718, 184064, 143119, 136073, 165003]


plt.plot(month_number, new_deaths)
plt.title('New Reported Deaths By Month (Globally)')
plt.xlabel('Month Number')
plt.ylabel('Number Of Deaths')
plt.show()
```

```python
def plot_cumulative_cases(country_name):
    country = who[who['Country'] == country_name]
    plt.plot(country['Date_reported'], country['Cumulative_cases'])
    plt.title('{}: Cumulative Reported Cases'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    plt.show()
plot_cumulative_cases('India')
```

```python
france = who[who['Country'] == 'France']
uk = who[who['Country'] == 'The United Kingdom']

plt.plot(france['Date_reported'], france['Cumulative_cases'],
         label='France')
plt.plot(uk['Date_reported'], uk['Cumulative_cases'],
         label='The UK')
plt.legend()
plt.show()
```

show method .show() allows us to control if we want plotted together or seperately 

```python
plt.plot(france['Date_reported'], france['Cumulative_cases'])
plt.title('France')
plt.show()

plt.plot(uk['Date_reported'], uk['Cumulative_cases'])
plt.title('The UK')
plt.show()
```

Comparing on same axes is more beneficial in this case as we see the Y range is almost 2x as high in the UK - judging by lines alone the growth rates look similar but the magnitude is much greater in the UK.

```python
france = who[who['Country'] == 'France']
uk = who[who['Country'] == 'The United Kingdom']
italy = who[who['Country'] == 'Italy']

plt.plot(france['Date_reported'], france['Cumulative_cases'],
         label='France')
plt.plot(uk['Date_reported'], uk['Cumulative_cases'],
         label='The UK')
plt.plot(italy['Date_reported'], italy['Cumulative_cases'],
         label='Italy')

plt.legend()
plt.show()
```
