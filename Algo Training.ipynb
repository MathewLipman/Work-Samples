```python
# Linear search algorithm
def linear_search(arr, target):
    # Iterate through the list
    for i, item in enumerate(arr):
        # If the target is found, return the index
        if item == target:
            return i
    # If the target is not found, return -1
    return -1

# Test the linear search algorithm
arr = [5, 3, 9, 1, 7, 6]
target = 7
index = linear_search(arr, target)
print(f"Index: {index}")
```

```python
# Binary search algorithm
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    # While the low index is less than or equal to the high index
    while low <= high:
        # Calculate the middle index
        mid = (low + high) // 2

        # If the target is found, return the index
        if arr[mid] == target:
            return mid
        # If the target is greater than the middle element, update the low index
        elif arr[mid] < target:
            low = mid + 1
        # If the target is smaller than the middle element, update the high index
        else:
            high = mid - 1

    # If the target is not found, return -1
    return -1

# Test the binary search algorithm
arr = [1, 3, 5, 6, 7, 9]
target = 7
index = binary_search(arr, target)
print(f"Index: {index}")
```

```python
# Bubble sort algorithm
def bubble_sort(arr):
    # Iterate through the list
    for i in range(len(arr)):
        # Iterate through the remaining unsorted elements
        for j in range(len(arr) - i - 1):
            # If the current element is greater than the next one, swap them
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Test the bubble sort algorithm
arr = [5, 3, 9, 1, 7, 6]
bubble_sort(arr)
print(f"Sorted array: {arr}")
```

```python
# Insertion sort algorithm
def insertion_sort(arr):
    # Iterate through the list, starting from the second element
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Move elements greater than the key to the right
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        # Insert the key in the correct position
        arr[j + 1] = key

# Test the insertion sort algorithm
arr = [5, 3, 9, 1, 7, 6]
insertion_sort(arr)
print(f"Sorted array: {arr}")
```

```python
# Partition function for quick sort
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    # Move the pivot element to its correct position
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# Quick sort algorithm
def quick_sort(arr, low, high):
    if low < high:
        # Find the partition index
        pi = partition(arr, low, high)

        # Recursively apply quick sort to the left and right partitions
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

# Test the quick sort algorithm
arr = [5, 3, 9, 1, 7, 6]
quick_sort(arr, 0, len(arr) - 1)
print(f"Sorted array: {arr}")
```
