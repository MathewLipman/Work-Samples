```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('/content/drive/MyDrive/Dataset/chinook.db')
```

```python
q1 = '''
SELECT name FROM sqlite_master WHERE type='table';
'''

df = pd.read_sql_query(q1, conn)
df
```

```python
q2 = ''' 
SELECT invoice_line.track_id, track.name, media_type.name, invoice_line.unit_price, invoice_line.quantity
FROM invoice_line
INNER JOIN track on track.track_id = invoice_line.track_id
INNER JOIN media_type on media_type.media_type_id = track.media_type_id
WHERE invoice_line.invoice_id = 4
'''
df = pd.read_sql_query(q2, conn)
df
```

```python
q3_1 = '''
SELECT * FROM invoice_line
WHERE invoice_id = 3;
'''

q3_2 = '''
SELECT * FROM invoice_line il
INNER JOIN track t ON t.track_id = il.track_id
WHERE invoice_id = 3;
'''

df1 = pd.read_sql_query(q3_1, conn)
df2 = pd.read_sql_query(q3_2, conn)

# two df solution
df1
df2
```

```python
q3 = '''
SELECT * 
FROM (
    SELECT * 
    FROM invoice_line
    WHERE invoice_id = 3
) il
INNER JOIN track t ON t.track_id = il.track_id;
'''

df = pd.read_sql_query(q3, conn)
df
```

```python
q3 = '''
WITH il AS (
    SELECT * 
    FROM invoice_line
    WHERE invoice_id = 3
)
SELECT * 
FROM il
INNER JOIN track t ON t.track_id = il.track_id;
'''

df = pd.read_sql_query(q3, conn)
df
```

```python
q4 = '''
SELECT invoice_line.track_id, 
       track.name AS track_name, 
       media_type.name AS track_type, 
       invoice_line.unit_price, 
       invoice_line.quantity
FROM invoice_line
INNER JOIN track ON track.track_id = invoice_line.track_id
INNER JOIN media_type ON media_type.media_type_id = track.media_type_id
WHERE invoice_line.invoice_id = 4;
'''

df = pd.read_sql_query(q4, conn)
df
```

```python
q5 = '''

SELECT invoice_line.track_id, 
       track.name AS track_name,
       artist.name AS artist_name,
       media_type.name AS track_type, 
       invoice_line.unit_price, 
       invoice_line.quantity
FROM invoice_line
INNER JOIN track ON track.track_id = invoice_line.track_id
INNER JOIN media_type ON media_type.media_type_id = track.media_type_id
INNER JOIN album ON track.album_id = album.album_id
INNER JOIN artist ON album.artist_id = artist.artist_id
WHERE artist_name = 'Green Day'

'''

df = pd.read_sql_query(q5, conn)
df
```
