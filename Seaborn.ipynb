SalePrice: price of the sale in USD

Gr Liv Area: above grade (ground) living area in square feet

Overall Qual: quality ratings of the overall material and finish of the house

Garage Area: garage area in square feet

Rooms : number of rooms

import seaborn library, standard alias in SNS

```python
import seaborn as sns
```

Built on matplotlib, Seaborn's replot parameters are data, x, y (and more)

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice')
plt.show()
```

When we use set_theme() method we remove the default matplotlib.pyplot template

```python
sns.set_theme()
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice')
plt.show()
```

setting hue to 'Overall Qual' in the replot parameter allows us to show the properties quality ratings related to materials and house finish with color - as it gets darker the quality rating increases and we can see the sales price does as well! The values in Overall Quality range from 1 to 10 making it an excellent choice for representation with Hue.

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual')
plt.show()
```

By setting the palette variable in our Seaborn (sns) relpot method we can really show case the changes in quality and intensity.

RdYlGn our argument in paletta contains, red yellow green but there are many different color palette examples 

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='RdYlGn')
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='seismic')
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='BrBG')
plt.show()
```

Adding the size parameter with the garage area variable allows us to pair the size displayed on the chart relative to the garage area of the properly

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area')
plt.show()
```

We can make the sizes more distinguishable by increasing the size range using sizes

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300))
plt.show()
```

We are now using size (only two options, 7 rooms or more, 6 rooms or less) to control the size of the the scatter marks representing homes on our seaborn plot
Notie we are adding the sizes using a list to map to the two values in 'Rooms'

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Rooms', sizes=[200,50])
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
             hue='Overall Qual', palette='RdYlGn', 
             size='Garage Area', sizes=(1,300))
plt.show()
```

Current properties we have represented in Seaborn (sns) - dataset, x coordinate, y coordinate, color, size

Shape is another property - currently all the graphs we have represented in Seaborn have used circles as **markers**, but there are more shapes.

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms')
plt.show()
```

Notice we added Rooms (two possible values) to our chart in the style parameter. We now have a cross representing 6 rooms or less and a circle representing 7 or more. 
Other shape options can be found here: https://matplotlib.org/3.3.3/api/markers_api.html

By using the marker parameter we can choose the different shapes that the data in the style parameter outputs

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', markers=['^','v'])
plt.show()
```

Variables we have chartered so far are:
**Gr Liv Area:** represented with x-coordinates.
**SalePrice:** represented with y-coordinates.
**Overal Qual:** represented with color.
**Garage Area:** represented with size.
**Rooms:** represented with shape.

We are now going to seperate our graph by plotting another variable, **Year:** which is the year when a property was built

the data in year is going to be output in the col parameter

It only contains two values - 1999 or older, 2000 or newer


```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', col='Year')
plt.show()
```
