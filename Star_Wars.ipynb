```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

```python
star_wars = pd.read_csv('/content/drive/MyDrive/Dataset/StarWars.csv', encoding="ISO-8859-1")
```

`RespondentID` — An anonymized ID for the respondent (person taking the survey)

`Gender` — the respondent's gender

`Age` — the respondent's age

`Household Income` — the respondent's income

`Education` — the respondent's education level

`Location `(Census Region) — the respondent's location

`Have you seen any of the 6 films in the Star Wars franchise? `— a Yes or No response

`Do you consider yourself to be a fan of the Star Wars film franchise?` — a Yes 
or No response 

```python
star_wars.columns
```

```python
star_wars.head(10)
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
print(star_wars.head(5))
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars['ranking_1'].replace('Star Wars: Episode I  The Phantom Menace', '1', inplace=True)
star_wars['ranking_2'].replace('Star Wars: Episode II  Attack of the Clones', '2', inplace=True)
star_wars['ranking_3'].replace('Star Wars: Episode III  Revenge of the Sith', '3', inplace=True)
star_wars['ranking_4'].replace('Star Wars: Episode IV  A New Hope', '4', inplace=True)
star_wars['ranking_5'].replace('Star Wars: Episode V The Empire Strikes Back', '5', inplace=True)
star_wars['ranking_6'].replace('Star Wars: Episode VI Return of the Jedi', '6', inplace=True)


```

```python
star_wars['ranking_1'].value_counts()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean(), label='male')
plt.legend()
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean(), label='female')
plt.legend()
plt.show()
```

```python
style.use('dark_background')
plt.bar(range(6), males[males.columns[3:9]].sum(), label='male')
plt.legend()
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum(), label='female')
plt.legend()
plt.show()
```
